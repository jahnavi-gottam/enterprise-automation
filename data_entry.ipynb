```python
import pandas as pd
import numpy as np
import joblib
import spacy
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
from fuzzywuzzy import fuzz
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load dataset
df = pd.read_csv(r"C:\Users\USER\Desktop\AutoGuide\data\User_Data.csv")

# Show first 5 rows
df.head()
```

```python
def store_user_data(user_id, name, email, phone, address, dob):
    new_entry = {
        "user_id": user_id,
        "name": name,
        "email": email,
        "phone": phone,
        "address": address,
        "dob": dob
    }
    
    # Append new data to DataFrame
    global df
    df = df.append(new_entry, ignore_index=True)
    
    # Save updated data
    df.to_csv("user_form_data.csv", index=False)
    print("User data saved successfully!")
```

```python
def autofill_form(user_id):
    # Check if user exists
    user_data = df[df["user_id"] == user_id]
    
    if not user_data.empty:
        print("ðŸ”¹ Auto-filling form for user:", user_id)
        return user_data.iloc[0].to_dict()
    else:
        print("âš  No previous data found for this user.")
        return None
```

```python
def autofill_form(user_id):
    if "user_id" not in df.columns:
        print("âš  Error: 'user_id' column missing in dataset.")
        return None

    # Ensure user_id matches stored format
    user_id = str(user_id)

    # Check if user_id exists
    user_data = df[df["user_id"] == user_id]

    if not user_data.empty:
        return user_data.iloc[0].to_dict()
    else:
        print("âš  No previous data found for this user.")
        return None
```

```python
while True:
    user_id = input("Enter User ID (or type 'exit' to stop): ")
    if user_id.lower() == 'exit':
        print("Automation ended.")
        break
    
    # Try auto-filling form
    form_data = autofill_form(user_id)
    
    if form_data:
        print(f"ðŸ”¹ Auto-Filled Form:\n{form_data}")
    else:
        print("âš  Please enter details manually.")
        
        name = input("Enter Name: ")
        email = input("Enter Email: ")
        phone = input("Enter Phone: ")
        address = input("Enter Address: ")
        dob = input("Enter Date of Birth (YYYY-MM-DD): ")
        
        store_user_data(user_id, name, email, phone, address, dob)
```
